```python
import pandas as pd
```

```python
first_half = pd.read_json("data/first_half_cards.json")
second_half = pd.read_json("data/second_half_cards.json")
```

```python
cards = pd.concat([first_half, second_half])
```

```python
cards = cards[['name', 'layout', 'mana_cost', 'cmc', 'type_line', 'oracle_text', 'colors',
       'color_identity', 'keywords', 'produced_mana', 'game_changer', 'set', 'set_name', 
       'rarity', 'power', 'toughness', 'flavor_text']]
cards.shape
```

```python
sorted_cards = cards.sort_values(by = 'name', ignore_index=True).drop_duplicates(subset = 'name')
sorted_cards.head()
```

```python
sorted_cards.to_csv('data/cleaned_cards.csv')
```
